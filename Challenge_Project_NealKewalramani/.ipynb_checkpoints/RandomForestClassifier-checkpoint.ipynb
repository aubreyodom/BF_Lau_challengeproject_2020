```python
import pandas as pd
import os
import shutil
from sklearn.ensemble import RandomForestClassifier
import getFeatures as gf
import numpy as np
from sklearn import metrics
import scipy
from matplotlib import pyplot as plt
from sklearn.model_selection import GridSearchCV
```

```python
train = pd.read_csv('train_set.csv')
test = pd.read_csv('test_set.csv')
ranks = pd.read_excel('ranked_curves.xlsx')
print(train)
```

```python
true_file2rank = dict()
for i in range(ranks.shape[0]):
    true_file2rank[ranks.loc[i, 'name']] = ranks.loc[i, 'rank']

train_file2rank = dict()
print('Train:')
for i in range(train.shape[0]):
    if train.iloc[i, 0] in true_file2rank.keys():
        train_file2rank[train.iloc[i, 0]] = true_file2rank[train.iloc[i, 0]]
    else:
        print(train.iloc[i, 0])

test_file2rank = dict()
print('Test:')
for i in range(test.shape[0]):
    if test.iloc[i, 0] in true_file2rank.keys():
        test_file2rank[test.iloc[i, 0]] = true_file2rank[test.iloc[i, 0]]
    else:
        print(test.iloc[i, 0])
```

```python
def get_all_files(directory):
    filenames = []
    for filename in os.listdir(directory):
        if filename.endswith('.xlsx') or filename.endswith('.xls'):
            filenames.append(directory+filename)
    return filenames

dataFiles = ['AeAeg2', 'AeAlbo2', 'AnGam2', 'CuQuin2', 'Dmel2']
for directory in dataFiles:
    filenames = get_all_files(directory)
    for filename in filenames:
        if filename in train_file2rank.keys():
            shutil.copy(directory + '/' + filename, 'train')
        elif filename in test_file2rank.keys():
            shutil.copy(directory + '/' + filename, 'test')
```

```python
print(len(train_file2rank.keys()))
print(len(test_file2rank.keys()))
```

```python
def reorder(d):
    dc = d.copy()
    
    filenames = d.keys()
    ranks = d.values()
    
    filenames = [i for i, _ in sorted(zip(filenames, ranks))]
    
    for i in range(len(filenames)):
        dc[filenames[i]] = i+1
    
    return dc

print(reorder(train_file2rank))
print(reorder(test_file2rank))
```

```python
train_file2rank = reorder(train_file2rank)
test_file2rank = reorder(test_file2rank) 

trainList = []
testList = []
for filename in train_file2rank.keys():
    l = list(gf.classify(filename, 'train/'))
    l.append(train_file2rank[filename])
    trainList.append(l)

for filename in test_file2rank.keys():
    l = list(gf.classify(filename, 'test/'))
    l.append(test_file2rank[filename])
    testList.append(l)
```

```python
columns = ['ratio', 'ideal', 'ran', 'sd', 'area_under_peaks', 'aup_normed', 'smoothed_error', 'rank']
trainDf = pd.DataFrame(trainList, columns = columns)
testDf = pd.DataFrame(testList, columns = columns)
```

```python
def find_min_index(l):
    minVal = np.inf
    minIndex = None
    
    for i in range(len(l)):
        if l[i] < minVal:
            minVal = l[i]
            minIndex = i
    return minIndex

def real_ranks(y_pred):
    y = list(y_pred)
    ranks = [0 for i in range(len(y))]
    
    for i in range(len(y)):
        minIndex = find_min_index(y)
        y[minIndex] = np.inf
        ranks[minIndex] = i+1
    
    return np.array(ranks)

x_train = trainDf[columns[:-1]]
y_train = trainDf[columns[-1]]
x_test = testDf[columns[:-1]]
y_test = testDf[columns[-1]]

clf = RandomForestClassifier()
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)

y_pred = real_ranks(y_pred)
print('Prediction:')
print(y_pred)
print('Test:')
print(np.array(y_test))
#print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

print(scipy.stats.wilcoxon(y_test, y_pred))

y_train_pred = clf.predict(x_train)
y_train_pred = real_ranks(y_train_pred)
print('Prediction:')
print(y_train_pred)
print('Train:')
print(np.array(y_train))
#scipy.stats.wilcoxon(y_train_pred, y_train)
print("Accuracy:",metrics.accuracy_score(y_train_pred, y_train))
```

```python
pVals = []
for i in range(10, 101):
    clf = RandomForestClassifier(n_estimators = i)
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    y_pred = real_ranks(y_pred)
    #print(i, "Accuracy:",metrics.accuracy_score(y_test, y_pred))
    pVals.append(scipy.stats.wilcoxon(y_test, y_pred)[1])
x = list(range(10, 101))
plt.plot(x, pVals)
plt.xlabel('n_estimators')
plt.ylabel('P value')
plt.show()
```

```python
def classify(directories):
    
    allFilenames = []
    for directory in directories:
        allFilenames += get_all_files(directory)
    
    file2feature = dict()
    for filename in allFilenames:
        l = list(gf.classify(filename, 'train/'))
        l.append(train_file2rank[filename])
        file2feature[filename] = l
```
