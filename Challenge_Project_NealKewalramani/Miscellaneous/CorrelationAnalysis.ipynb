```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_excel('Correlation.xls')
heatdf = df.iloc[:, 1:]
corrdf = heatdf.corr()
print(corrdf)
ax = sns.heatmap(corrdf)
plt.title('Correlation between each feature')
plt.savefig('HeatmapCorrelation.pdf')
plt.show()
```
